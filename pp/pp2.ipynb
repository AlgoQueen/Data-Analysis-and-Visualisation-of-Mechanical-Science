```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values, avoiding points where tan(x) and sec(x) are undefined
x = np.linspace(-np.pi/2 + 0.1, np.pi/2 - 0.1, 400)

# Define the functions
y = np.tan(x)
y_prime = 1 / np.cos(x)**2  # sec^2(x)
y_double_prime = 2 * y_prime * y  # 2 * sec^2(x) * tan(x)

# List of functions, labels, and line styles for looping
functions = [y, y_prime, y_double_prime]
titles = [r"$y = \tan(x)$", r"$y' = \sec^2(x)$", r"$y'' = 2\sec^2(x)\tan(x)$"]
colors = ['blue', 'green', 'red']
linestyles = ['-', '--', '-.']

# Create a figure and subplots
plt.figure(figsize=(10, 8))

# Loop to plot each function in a subplot
for i, (func, title, color, linestyle) in enumerate(zip(functions, titles, colors, linestyles)):
    plt.subplot(3, 1, i + 1)
    plt.plot(x, func, color=color, linestyle=linestyle, label=title)
    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x, y) = x^2 + y^2


def f(x, y):
    return x**2 + y**2

# Define the gradient of f(x, y)


def gradient(x, y):
    df_dx = 2 * x
    df_dy = 2 * y
    return np.array([df_dx, df_dy])

# Steepest Descent Method


def steepest_descent(initial_point, alpha, tolerance, max_iterations=1000):
    x, y = initial_point
    points = [(x, y)]  # To store the path of the optimization
    for iteration in range(max_iterations):
        grad = gradient(x, y)
        new_x = x - alpha * grad[0]
        new_y = y - alpha * grad[1]
        points.append((new_x, new_y))

        # Check for convergence
        if np.linalg.norm([new_x - x, new_y - y]) < tolerance:
            print(f"Converged in {iteration + 1} iterations.")
            break

        x, y = new_x, new_y
    else:
        print("Max iterations reached without convergence.")

    return x, y, points

# Plotting the contour and the path of steepest descent


def plot_contour_and_path(points):
    # Create a grid of points
    x = np.linspace(-1.5, 1.5, 400)
    y = np.linspace(-1.5, 1.5, 400)
    X, Y = np.meshgrid(x, y)
    Z = f(X, Y)

    # Create the contour plot
    plt.figure(figsize=(8, 6))
    plt.contour(X, Y, Z, levels=20, cmap='viridis')
    plt.colorbar(label='f(x, y)')

    # Plot the path
    path = np.array(points)
    plt.plot(path[:, 0], path[:, 1], 'ro--', label='Steepest Descent Path')
    plt.plot(path[0, 0], path[0, 1], 'go', label='Start Point')
    plt.plot(path[-1, 0], path[-1, 1], 'bo', label='End Point')

    plt.title('Contour Plot with Steepest Descent Path')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.show()


# Initialize parameters
initial_point = (1, 1)
alpha = 0.1
tolerance = 1e-6

# Run the steepest descent method
min_x, min_y, points = steepest_descent(initial_point, alpha, tolerance)
print(f"Minimum point: ({min_x}, {min_y})")

# Plot the contour and the optimization path
plot_contour_and_path(points)
```
