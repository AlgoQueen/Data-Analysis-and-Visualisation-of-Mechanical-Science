```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x,y):
    return x**2 + y**2
def grad_f(x,y):
    df_dx=2*x
    df_dy=2*y
    return np.array([df_dx, df_dy])
```

```python
#sanity check
print(f(1,1))
print(grad_f(1,1))
```

```python
x,y =[1,1] #starting point
Nmax=20
alpha = 0.1
for i in np.arange(Nmax):
    dx,dy = grad_f(x,y)
    x = x- alpha*dx
    y = y- alpha*dy# modify the code to save the values in an array
    plt.plot(x,y,'-ro')
    # print(x,y)
```

```python
x1 = np.linspace(-2,2)
y1 = np.linspace(-2,2)
[X,Y]= np.meshgrid(x1,y1)
Z = f(X,Y)
# plt.pcolor(X,Y,Z)
# plt.contour(X,Y,Z,20)
plt.contour(X, Y,Z,[0.5])
plt.contour(X, Y,Z,0.1 +np.arange(10)*0.1)

x, y = [1, 1]  # starting point
Nmax = 20
alpha = 0.1
for i in np.arange(Nmax):
    dx, dy = grad_f(x, y)
    x = x - alpha*dx
    y = y - alpha*dy  # modify the code to save the values in an array
    plt.plot(x, y, '-ro')
    # print(x,y)
```

# Interpolation and Extrapolation

## Interpolation


```python
from scipy.interpolate import interp1d
```

```python
x = np.array([0,1,2,3,4,6]) 
y= np.cos(x)
plt.plot(x,y,'-ok')
xi =0.5 + np.arange(0,6)
# lf = interp1d(x,y, 'linear')
# lf = interp1d(x, y, 'cubic')
# lf = interp1d(x, y, 'nearest') # nearest neighbor interpolation fastest, cubic is slowest
#spline interpolation
lf = interp1d(x, y, 'linear')
yi = lf(xi)
plt.plot(xi,yi,'sr')
```

```python
from scipy.interpolate import griddata
x =np.linspace(-3.5,3.5,20)
y = np.linspace(-3.5, 3.5, 20)
[X,Y]= np.meshgrid(x,y) #create a grid
Z = np.cos(X)*np.sin(Y)
# print(Z)
plt.pcolor(X,Y,Z) #bright maximas, dark minimas
xi = np.linspace(-3.5,3.5,20)
yi = np.linspace(-3.5, 3.5, 20)
[Xi,Yi]= np.meshgrid(xi,yi)
xi = 0.2 ; yi = 0.5
zi = griddata((np.ndarray.flatten(X),np.ndarray.flatten(Y)),(np.ndarray.flatten(Z)),(xi,yi))    
```

```python
plt.pcolor(Xi, Yi, zi)
```
