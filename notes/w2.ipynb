```python
markers = ['x', 's', '^','o','v','d']
from itertools import cycle
markercycle = cycle(markers) # cycle is a generator that will keep on generating the next element in the list
plt.figure(figsize=(6,4))
x = np.linspace(0,10,10)
y_a = [x,x**2, np.sqrt(x), np.log(x+1)]

for y in y_a:
    plt.plot(x,y, marker=next(markercycle), label=f'y ={y}')
    # plt.plot(x, y, marker=next(markercycle), label=next(np.array(y_a)))
    plt.legend()
```

```python
#plotting Simple Harmonic Motion
phi = 0;
A = 1;
omega = 20; 
T = 2*np.pi/omega
N=100;
t= np.linspace(0,2.5*T,N)

def displacement(t, A, phi, omega):
    return A*np.cos(omega*t +phi)

def velocity(t,A,phi,omega):
    return -A*omega*np.sin(omega*t + phi)

def acc(t,A,phi,omega):
    return -A*np.sin(omega*t + phi)*omega**2


plt.plot(t, velocity(t, A, phi, omega))
```

```python
N_max = 10
alpha =0.5

def y(x):
    return (x-2)**2

def yprime(x):
    return 2*(x-2)

x = 420*np.ones(N_max)
x[0] = 3
err =1
err_threshold = 1e-3
for n in np.arange(N_max-1):
    x[n+1] =x [n] - alpha*yprime(x[n])
    err =( 1-np.abs(x[n+1]/x[n]))
    if err<err_threshold:
        print(err_threshold)
        print('Error threshold reached')
        break
print('Converged Solution is',x[n+1])
print(n)
```

```python
x_iters = x[0:n+1]
y_iters = y(x_iters)
x = np.linspace(1,3.5);
y=y(x_iters)
plt.plot(x,y(x))
plt.plot(x_iters,y_iters)
```
