```python
import numpy as np
import matplotlib.pyplot as plt

def f(x,y):
    return x**2 -x*y + y**2

def grad_f(x,y):
    return [2*x - y, -x + 2*y]

x,y = 1,1
alpha =0.1
tol=1e-6
max_iter = 1000


points = [(x, y)]  # To store the path of the optimization
for i in range(max_iter):
    grad = grad_f(x, y)
    new_x = x - alpha * grad[0]
    new_y = y - alpha * grad[1]
    points.append((new_x, new_y))

        # Check for convergence
    if np.linalg.norm([new_x - x, new_y - y]) < tol:
        print(f"Converged in {i + 1} iterations.")
        break

    x, y = new_x, new_y
    
print(f"Minimum point: ({x:.6f}, {y:.6f})")

print(f"Minimum value: {f(x,y):.6f}")

#plotting
x_vals = np.linspace(-2,2,400)
y_vals = np.linspace(-2,2,400)
X,Y = np.meshgrid(x_vals,y_vals)
Z = f(X,Y)

plt.figure(figsize=(10,6))
contour=plt.contour(X,Y,Z,levels=50, cmap='viridis',   alpha=0.6) # alpha is the transparency
plt.colorbar(contour, label='Function value')

points = np.array(points)
plt.plot(points[:,0], points[:,1], 'r-o',markersize=3, label='Steepest Descent Path')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Steepest Descent Method')
```
