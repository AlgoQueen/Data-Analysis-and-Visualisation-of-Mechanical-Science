```python
import numpy as np
import matplotlib.pyplot as plt

rho = 1000  # kg/m^3
C = 2000  # J/K
A= 0.5  # m^2
V = 0.25  # m^3
t = np.linspace(0, 120, 121)  # Time from 0 to 120 seconds with 1 second intervals

T0 = 80   # Initial temperature in Kelvin
T_amb = 20   # Ambient temperature in Kelvin

# Heat transfer coefficients
h = [5, 10, 15]  # W/m^2 K data type: list

def temperature(t, h):
    return T_amb + (T0 - T_amb) * np.exp((-h * A * t )/ (rho * V * C))


linestyles = ['-', '--', '-.']
colors = ['r', 'g', 'b']
markers = ['o', 's', 'd']

plt.figure(figsize=(10, 6))

for i, h_i in enumerate(h):
        print(f'Heat transfer coefficient: {h_i} W/m^2 K')
        T=[];
        T = temperature(t, h_i)
        print(f'Temperature is {T}')
        plt.plot(t, T, label=f'h = {h_i} $W/m^2 K$',linestyle=linestyles[i], color=colors[i], marker=markers[i], markevery=10)

plt.xticks(np.arange(0, 121, 10))

plt.xlabel('Time (s)')
plt.ylabel('Temperature (°C)')
plt.title('Temperature vs Time')
# add small grids
plt.grid(True, which='both', linestyle='-', linewidth=0.5)
plt.legend()
```
